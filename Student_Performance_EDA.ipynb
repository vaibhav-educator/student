# ðŸ§  Student Performance EDA with Python
In this notebook, we'll perform data cleaning, transformation, and basic EDA on the Student Performance dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
```

```python
# Load the dataset
df = pd.read_csv('student_data.csv')
df.head()
```

```python
# Basic info
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics
df.describe()
```

```python
# Distribution of Total Score
sns.histplot(df['Total Score'], bins=10, kde=True)
plt.title('Total Score Distribution')
plt.show()
```

```python
# Average Total Score by Department
avg_scores = df.groupby('Department')['Total Score'].mean()
print(avg_scores)

avg_scores.plot(kind='bar', title='Average Total Score by Department')
plt.ylabel('Average Score')
plt.show()
```

```python
# Students with Attendance below 75%
low_attendance = df[df['Attendance (%)'] < 75]
low_attendance
```

```python
# Assign Grades based on Total Score
def assign_grade(score):
    if score >= 85:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 50:
        return 'C'
    else:
        return 'D'


df['Grade'] = df['Total Score'].apply(assign_grade)
df[['Name', 'Total Score', 'Grade']].head()
```
